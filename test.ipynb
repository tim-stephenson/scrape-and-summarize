```python
from json import dumps
import requests
import bs4
```

```python



r = requests.post('https://www.quantamagazine.org/graphql', json=  {
	"operationName": "getPopular",
	"variables": {
		"cat": "189",
		"limit": 6,
	},
	"query": """query getPopular($cat: String, $author : Int, $post_type : String, $limit : Int, $range : String, $exclude_ids : String) {
  popular: getPopular(cat: $cat , author: $author , post_type: $post_type , limit: $limit , range: $range , exclude_ids: $exclude_ids ) {
    ... on Post {
      id
      title
      link
      authors {
        name
        link
        __typename
      }
      featured_media_image {
        ...ImageFields
        __typename
      }
      acf {
        featured_block_title
        featured_image_default {
          ...ImageFields
          __typename
        }
        featured_image_square {
          ...ImageFields
          __typename
        }
        __typename
      }
      __typename
    }
    __typename
  }
}

fragment ImageFields on Image {
  alt
  caption
  url
  width
  height
  sizes {
    thumbnail
    square_small
    square_large
    medium
    medium_large
    large
    __typename
  }
  __typename
}"""
} )
r.status_code
```

```python
r = requests.post('https://www.quantamagazine.org/graphql', json=  {
	"operationName": "getPopular",
	"variables": {
		"cat": "189",
		"limit": 1,
	},
	"query": """query getPopular($cat: String, $author : Int, $post_type : String, $limit : Int, $range : String, $exclude_ids : String) {
  popular: getPopular(cat: $cat , author: $author , post_type: $post_type , limit: $limit , range: $range , exclude_ids: $exclude_ids ) {
    ... on Post {
      id
      title
      link
      content
      authors {
        name
        link
      }
      acf {
        featured_block_title
        featured_image_default {
          ...ImageFields
        }
        featured_image_square {
          ...ImageFields
        }
      }
    }
  }
}

fragment ImageFields on Image {
  alt
  caption
  url
  width
  height
  sizes {
    thumbnail
    square_small
    square_large
    medium
    medium_large
    large
  }
}"""
} )
r.raise_for_status()
print(dumps(r.json(), indent=4 ) )
```
